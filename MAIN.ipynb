The datasets is the user behaviour dataset in mobile phones from Kaggle

```python
import pandas as pd#handling the dataset
file_path=pd.read_csv("/content/user_behavior_dataset.csv")
print(file_path.head())
print(file_path.tail())
```

```python
file_path.fillna(file_path.select_dtypes(include="number").mean())
```

```python
file_path.drop(columns=["User Behavior Class"])
```

```python
file_path.rename(columns={"Device Model":"Device","User ID":"ID"})
```

```python
file_path.drop_duplicates(inplace=True)
```

```python
file_path.select_dtypes(include="number").mean()
```

```python
file_path.select_dtypes(include="number").median()#converts into ascending order and takes the middle value
```

```python
file_path.select_dtypes(include="number").corr()#to compute the correlation of a dataframe(helps stong corrlation closer to 1 or -1)
```

```python
file_path.select_dtypes(include="number").std()#it measures the amount of varation in a dataset,higher std means the values are more spread out
```

```python
file_path.drop('User Behavior Class',axis=1,inplace=True)
```

```python
file_path.to_csv('clean.csv',index=False)
```

```python
import seaborn as sns#for visualization
import matplotlib.pyplot as pl#for plotting
import pandas as pd#for handling the datasets
file=pd.read_csv("/content/clean.csv")
pl.figure(figsize=(10,8))#sets the figsize to 10x8 inches
corr=file.corr(numeric_only=True)#computes the corrleation matrix for the numerical columns
sns.heatmap(corr,annot=True,cmap='coolwarm',linewidth=1)#linewidth sets the grid lines#displays correlation values inside the cells
pl.show()
#Correlation of HeatMap
```

```python
#Line Chart
pl.figure(figsize=(10,7))
pl.plot(file['Device Model'],file['Age'],marker='o',linestyle='--',linewidth=2)
pl.title("Line Chart")
pl.xlabel("Model")
pl.ylabel("Age")
pl.show()
```

```python
#Histogram(used for comparing numerical data)
pl.hist(file['Age'],bins=5,color='red',edgecolor='black')
pl.title("Histogram")
pl.xlabel("Model")
pl.ylabel("Age")
pl.show()
```

```python
#Scatter Plot
pl.scatter(file['Age'],file['App Usage Time (min/day)'],cmap='coolwarm')
pl.title("Scatter plot")
pl.xlabel("Age")
pl.ylabel("App Usage (min/day)")
pl.show()
```

```python
#BAR PLOT(used for comparing categorical data)
pl.bar(file['Device Model'],file['Screen On Time (hours/day)'],color='green')
pl.title("Bar Plot")
pl.xlabel("Model")
pl.ylabel("Screen On Time (hours/day)")
pl.show()
```

```python
#PIE CHART
labels = file['Device Model'].unique()
sizes = file.groupby('Device Model')['Screen On Time (hours/day)']. sum()
pl.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=60)
pl.title('Distribution of Device Model Usage')
pl.axis('equal')
pl.show()
```
